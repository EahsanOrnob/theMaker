```python
import numpy as np
import pandas as pd
import sklearn

import time
from sklearn import linear_model
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from sklearn import metrics
from sklearn.metrics import accuracy_score
encode = LabelEncoder() 
```

```python
data=pd.read_csv('data.csv')
data.describe()

print(data)
```

```python
data.head()
```

```python
x=data.iloc[:,0:16].values
y=data.iloc[:,16].values

x[:,14] = encode.fit_transform(x[:,14])
x[:,15] = encode.fit_transform(x[:,15])
print(x)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=.50,random_state=200)
print("Train = ",len(X_train))
print("Test = ",len(X_test))
```

```python
#Implementing Naive Bayes 

start_time=time.time()
GaussNb=GaussianNB()
GaussNb.fit(X_train,Y_train)

y_pred=GaussNb.predict(X_test)
accuracy= accuracy_score(Y_test,y_pred)*100
print("Execution time: ",time.time()-start_time)
print("Test Accuracy: ",round(accuracy,1),"%")
```

```python
#Implementing Decision Tree
start_time=time.time()
DecisionTree=DecisionTreeClassifier(random_state=200)
DecisionTree.fit(X_train,Y_train)
```

```python
y_prede=DecisionTree.predict(X_test)
accuracy=accuracy_score(Y_test,y_prede)*100
print("Execution time: ",time.time()-start_time)
print("Test Accuracy: ",round(accuracy,1),"%")
```

```python
#Implementing Neural network
start_time=time.time()
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,  hidden_layer_sizes=(10, 5), random_state=200)
clf.fit(X_train,Y_train)
```

```python
y_pred=clf.predict(X_test)
accuracy=accuracy_score(Y_test,y_pred)*100
print("Execution time: ",time.time()-start_time)
print("Test Accuracy: ",round(accuracy,1),"%")
```
